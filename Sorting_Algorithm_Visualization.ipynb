# Scatter Plot

```python
import matplotlib.pyplot as plt
test = "normal"
idx_df = df.set_index(["Algo","Test"])
ax = idx_df.loc["bubble",test].plot(kind="scatter",x="n",y="TimeUsedMS",label="bubble",color="Blue")
idx_df.loc["insertion",test].plot.scatter(x="n",y="TimeUsedMS",label="insertion",ax=ax,color="Red")
idx_df.loc["selection",test].plot.scatter(x="n",y="TimeUsedMS",label="selection",ax=ax,color="Green")
idx_df.loc["merge",test].plot.scatter(x="n",y="TimeUsedMS",label="merge",ax=ax,color="Orange")
idx_df.loc["quick",test].plot.scatter(x="n",y="TimeUsedMS",label="quick",ax=ax,color="Grey")
idx_df.loc["flagged_bubble",test].plot.scatter(x="n",y="TimeUsedMS",label="flagged_bubble",ax=ax,color="Pink")
plt.title(f"Sorting Algorithms vs {test} cases")
plt.show()
```
